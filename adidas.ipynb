# I. Giới thiệu về tập dữ liệu

**1. Nguồn gốc**

**2. Ý nghĩa của tập dữ liệu**

* Units Sold: Số lượng sản phẩm đã bán.
* Total Sales: Tổng doanh thu từ việc bán sản phẩm.
* Operating Profit: Lợi nhuận vận hành thu được từ sản phẩm.
* Operating Margin: Biên lợi nhuận vận hành.
* Sales Method: Hình thức bán hàng (ví dụ: In-store, Online).

# Một số phương pháp phát hiện giá trị ngoại lai trong tập dữ liệu Adidas
**1. Phương pháp IQR (Interquartile Range)**

* Xác định ngoại lai dựa trên khoảng tứ phân vị: Giá trị nằm ngoài khoảng (Q1 - 1.5IQR, Q3 + 1.5IQR) được coi là ngoại lai.
* Ứng dụng: Phát hiện giá trị bất thường trong "Price per Unit" hoặc "Total Sales".
* Ưu điểm: Dễ thực hiện, phù hợp với dữ liệu có phân phối chuẩn.

**2. Phương pháp Percentile Capping**

* Cắt bỏ các giá trị nằm ngoài phạm vi 1% - 99% để hạn chế ảnh hưởng của ngoại lai.
* Ứng dụng: Điều chỉnh giá trị bất thường trong "Units Sold" hoặc "Operating Profit".
* Ưu điểm: Giữ lại nhiều dữ liệu hơn so với IQR, tránh loại bỏ quá nhiều điểm dữ liệu quan trọng.

```python
# Kiểm tra các chỉ số của cột Units Sold
print(df['Units Sold'].describe())
```

```python
# xác định giá trị ngoại lai

# Tính các giá trị Q1, Q3 và IQR cho "Units Sold"
Q1 = df['Units Sold'].quantile(0.25)  # Phân vị thứ 25%
Q3 = df['Units Sold'].quantile(0.75)  # Phân vị thứ 75%
IQR = Q3 - Q1  # Khoảng tứ phân vị

# Xác định ngưỡng dưới và ngưỡng trên
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Lọc dữ liệu, chỉ giữ các giá trị hợp lệ
df_cleaned = df[(df['Units Sold'] >= lower_bound) & (df['Units Sold'] <= upper_bound)]

# Kiểm tra số dòng sau khi xử lý
print(f"Số dòng trước khi xử lý: {len(df)}")
print(f"Số dòng sau khi loại bỏ ngoại lai cột 'Units Sold': {len(df_cleaned)}")
```
