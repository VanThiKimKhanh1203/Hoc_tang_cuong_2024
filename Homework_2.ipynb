```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lớp mô phỏng trò chơi
class GameSimulation:
    def __init__(self, p, i, N, T):
        self.p = p  # xác suất thắng
        self.q = 1 - p  # xác suất thua
        self.i = i  # số tiền ban đầu
        self.N = N  # số tiền mục tiêu
        self.T = T  # số lượt chơi tối đa

    def play_game(self):
        balance = self.i
        rounds = 0
        history = [balance]

        while balance > 0 and balance < self.N and rounds < self.T:
            if np.random.rand() < self.p:
                balance += 1  # thắng 1 USD
            else:
                balance -= 1  # thua 1 USD
            history.append(balance)
            rounds += 1

        return history
```

```python
# Hàm để mô phỏng nhiều lần
def simulate_games(num_simulations, p, i, N, T):
    results = []
    for _ in range(num_simulations):
        game = GameSimulation(p, i, N, T)
        results.append(game.play_game())
    return results
```

```python
# Hàm tính PMF
def calculate_pmf(results, N):
    final_balances = [result[-1] for result in results]
    pmf = np.zeros(N+1)
    for balance in final_balances:
        pmf[balance] += 1
    pmf /= len(results)  # chia cho tổng số lần mô phỏng
    return pmf

# Hàm tính CDF
def calculate_cdf(pmf):
    cdf = np.cumsum(pmf)  # tính tổng tích lũy
    return cdf
```

```python
# Hàm vẽ PMF và CDF
def plot_pmf_cdf(pmf, cdf, N):
    x = np.arange(N+1)

    plt.figure(figsize=(10, 6))

    # Vẽ PMF
    plt.subplot(1, 2, 1)
    plt.bar(x, pmf)
    plt.title('PMF')
    plt.xlabel('Số tiền')
    plt.ylabel('Xác suất')

    # Vẽ CDF
    plt.subplot(1, 2, 2)
    plt.plot(x, cdf, drawstyle='steps-post')
    plt.title('CDF')
    plt.xlabel('Số tiền')
    plt.ylabel('Xác suất tích lũy')

    plt.tight_layout()
    plt.show()
```

```python
# Vẽ đồ thị
def plot_simulation(results, title):
    plt.figure(figsize=(10, 6))
    for result in results:
        plt.plot(result)
    plt.title(title)
    plt.xlabel('Số lượt chơi')
    plt.ylabel('Số tiền còn lại')
    plt.show()

# Thông số ban đầu
p = 0.25  # xác suất thắng
i = 20   # số tiền ban đầu
N = 50   # mục tiêu
T = 1000  # số lượt chơi tối đa
num_simulations = 10  # số lần mô phỏng

# Mô phỏng và vẽ kết quả
results = simulate_games(num_simulations, p, i, N, T)
plot_simulation(results, f'mô phỏng trò chơi với p={p}, i={i}, N={N}, T={T}')
```

```python
# Tính và vẽ PMF và CDF từ kết quả mô phỏng
pmf = calculate_pmf(results, N)
cdf = calculate_cdf(pmf)
plot_pmf_cdf(pmf, cdf, N)
```
